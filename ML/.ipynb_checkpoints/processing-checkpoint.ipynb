```python
import cv2
import mediapipe as mp
import json
import os
from glob import glob
from PIL import Image
import numpy as np

# Mediapipe Pose ëª¨ë¸ ë¡œë“œ
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# JSON ë°ì´í„° êµ¬ì¡° ì´ˆê¸°í™”
data = {
    "frames": [],
    "type": "561",
    "type_info": {
        "key": "561",
        "type": "ë§¨ëª¸ ìš´ë™",
        "pose": "ì—ë“œë¦°ìì„¸",
        "exercise": "í‘¸ì‹œì—…",
        "conditions": [
            {"condition": "ì²™ì¶”ì˜ ì¤‘ë¦½", "value": True},
            {"condition": "ì´ì™„ì‹œ íŒ”ê¿ˆì¹˜ 90ë„", "value": True},
            {"condition": "ê°€ìŠ´ì˜ ì¶©ë¶„í•œ ì´ë™", "value": True},
            {"condition": "ì†ì˜ ìœ„ì¹˜ ê°€ìŠ´ ì¤‘ì•™ ì—¬ë¶€", "value": True},
            {"condition": "ê³ ê°œ ì –í˜/ìˆ™ì„ ì—¬ë¶€", "value": True}
        ],
        "description": "ì •ìì„¸"
    }
}

# ì´ë¯¸ì§€ í´ë” ì§€ì •
image_folder = "D:/Studying/gradu/013.í”¼íŠ¸ë‹ˆìŠ¤ìì„¸/1.Training/ì›ì‹œë°ì´í„°/body_09/Day35_201107_F/1/A/561-1-3-27-Z18_A"  # ì´ë¯¸ì§€ê°€ ë“¤ì–´ìˆëŠ” í´ë” ê²½ë¡œ
image_files = sorted(glob(os.path.join(image_folder, "*.jpg")))  # ì´ë¯¸ì§€ í™•ì¥ì í™•ì¸ í›„ ë³€ê²½ ê°€ëŠ¥

# Mediapipe Pose ëª¨ë¸ ì‹¤í–‰
with mp_pose.Pose(static_image_mode=True, model_complexity=2, enable_segmentation=False) as pose:
    for img_file in image_files:
        # í™€ìˆ˜ ë²ˆì§¸ ì´ë¯¸ì§€ë§Œ ì²˜ë¦¬
        if int(img_file[-5]) % 2 != 1:
            continue

        print("ğŸ“‚ Loading image:", img_file)

        try:
            # PILì„ ì‚¬ìš©í•˜ì—¬ ì´ë¯¸ì§€ ë¡œë“œ
            image = Image.open(img_file).convert("RGB")  # RGB ë³€í™˜
            image_np = np.array(image)  # NumPy ë°°ì—´ë¡œ ë³€í™˜

            # ì´ë¯¸ì§€ í¬ê¸° ê°€ì ¸ì˜¤ê¸°
            image_width, image_height = image.size

            # Mediapipeì— ì…ë ¥í•  ìˆ˜ ìˆë„ë¡ ë°°ì—´ ë³€í™˜ (H, W, C ìˆœì„œ ìœ ì§€)
            results = pose.process(image_np)

            # í‚¤í¬ì¸íŠ¸ ì¢Œí‘œ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬ (í”½ì…€ ë‹¨ìœ„ ë³€í™˜)
            keypoints = {}

            if results.pose_landmarks:
                for idx, landmark in enumerate(results.pose_landmarks.landmark):
                    keypoints[f"Point_{idx}"] = {
                        "x": int(landmark.x * image_width),  # í”½ì…€ ì¢Œí‘œë¡œ ë³€í™˜
                        "y": int(landmark.y * image_height)  # í”½ì…€ ì¢Œí‘œë¡œ ë³€í™˜
                    }

            # JSON êµ¬ì¡°ì— ë°ì´í„° ì¶”ê°€
            frame_data = {
                "view1": {
                    "pts": keypoints,
                    "active": "Yes" if results.pose_landmarks else "No",
                    "img_key": img_file
                }
            }

            data["frames"].append(frame_data)

        except Exception as e:
            print(f"âŒ ì´ë¯¸ì§€ ì²˜ë¦¬ ì˜¤ë¥˜ ({img_file}): {e}")
            continue

# JSON íŒŒì¼ ì €ì¥
json_output_path = "D:/Studying/gradu/013.í”¼íŠ¸ë‹ˆìŠ¤ìì„¸/1.Training/ì›ì‹œë°ì´í„°/body_09/Day35_201107_F/1/A/561-1-3-27-Z18_A/output_pixel.json"
with open(json_output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

print(f"âœ… í”½ì…€ ì¢Œí‘œ JSON íŒŒì¼ì´ {json_output_path}ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
